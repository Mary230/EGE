Два игрока, Петя и Ваня, играют в следующую игру. Перед игроками лежит куча камней. Игроки ходят по очереди, первый ход делает Петя. За один ход игрок может

1.   **добавить в кучу один камень или**
2.   **добавить в кучу два камня или**
1.   **увеличить количество камней в куче в два раза**

*Например, имея кучу из 10 камней, за один ход можно получить кучу из 11, 12 или 20 камней. У каждого игрока, чтобы делать ходы, есть неограниченное количество камней.*

Игра завершается в тот момент, когда количество камней в куче **превышает 37**. Победителем считается игрок, сделавший последний ход, то есть первым получивший кучу, в которой будет 38 или больше камней.
**В начальный момент в куче было S камней, 1 ≤ S ≤ 37.**
Будем говорить, что игрок имеет выигрышную стратегию, если он может выиграть при любых ходах противника. Описать стратегию игрока  — значит, описать, какой ход он должен сделать в любой ситуации, которая ему может встретиться при различной игре противника. В описание выигрышной стратегии не следует включать ходы следующего стратегии игрока, которые не являются для него безусловно выигрышными.


В любой задаче на теорию игр нам понадобится функция, которая будет проверять, что позиция является выигрышной, то есть проверять что **по крайней мере один из возможных ходов приведет игрока в выигрыш:**

```python
def PW1(s):
    return s+1>=38 or s+2>=38 or s*2>=38
```

Тут нам понадобится вторая функция, которая будет проверять, что данная позиция проигрышная

**Если в задании нам говорят найти ВОЗМОЖНОЕ число S, при котором Петя НЕ выиграет свои первым ходом, а Ваня может выиграть своим ходом, то нужно проверить что данная позияи S:**


1.   Не является выигрышной
2.   Какой-то из из ходов из этой позиции приведет нас в выигрышную позицию


Далее буду называть это возможным прогрышем

```python
def PP1(s): 
    return PW1(s)!=True and (PW1(s+1) #PW1(s) должно быть ЛОЖЬ, тогда эта позиция Не выигрышная
                             or PW1(s+2) or PW1(s*2))  # Хотя бы ОДИН из ходов ведет нас в выигрышную позицию
```

**Если в задании нам говорят найти такое число S, что Петя проиграет своим первым ходом, а Ваня выиграет при любом ходе Пети, то нам нужно проверить, что данная позиция S:**


1.   Не является выигрышной
2.   Любой ход из этой позиции приведет нас в выигрышную позицию

Далее буду называть это гарантированным прогрышем

```python
def PP1(s): 
    return PW1(s)!=True and (PW1(s+1) #PW1(s) должно быть ЛОЖЬ, тогда эта позиция Не выигрышная
                             and PW1(s+2) and PW1(s*2))  # ВСЕ ходы из этой позиции ведут в выигрышную
```

Тогда нам нужно будет просто найти все числа, для которых функция  PP1() возвращает True
Сделаю это с помощью прохода по циклу для всех возможных значений S (беру это из задачи)


```python
# 1. Поменять условия выигрыша в PW1
# 2. Поменять операции исходя из ходов по задаче
# 3. Границы цикла
def PW1(s):
    return s+1>=38 or s+2>=38 or s*2>=38

def PP1(s): #ответ для 19ой с вопросом ВОЗМОЖНЫХ значений S
    return PW1(s)!=True and (PW1(s+1)
                             or PW1(s+2) or PW1(s*2))

for i in range(1,38):
    if PP1(i): print(i) #отсюда просто возьму мин. значение
```

— Петя не может выиграть за один ход;
— Петя может выиграть своим вторым ходом независимо от того, как будет ходить Ваня.

**Что тут?** Нам просто нужно найти три таких числа, которые ведут в нашу ГАРАНТИРОВАНО прогрышную стратегию
**Как?** Проверить что какая-то из операций с нашим числом ведет в прогрышную стратешию

```python
def PW2(s):
    return PP1(s+1) or PP1(s+2) or PP1(s*2) 
    # то есть просто берем какое-то число и проверяем, что какая-то операция с этим числом позволит превратить его в нашу проигрышную стратегию
```

**ВАЖНО**! 
**Тут мы используем функцию для гарантированного проигрыша (см. задачу 19)**

```python
# 1. Поменять условия выигрыша в PW1
# 2. Поменять операции исходя из ходов по задаче
# 3. Границы цикла
def PW1(s):
    return s+1>=38 or s+2>=38 or s*2>=38

def PP1(s):
    return PW1(s)!=True and (PW1(s+1) #вот тут как раз везде будет and для гарантии проигрыша
                             and PW1(s+2) and PW1(s*2))
    
def PW2(s):
    return PP1(s+1) or PP1(s+2) or PP1(s*2) 

for i in range(1,38):
    if PW2(i): print(i)
```

Найдите минимальное значение S, при котором одновременно выполняются два условия:
— у Вани есть выигрышная стратегия, позволяющая ему выиграть первым или вторым ходом при любой игре Пети;
— у Вани нет стратегии, которая позволит ему гарантированно выиграть первым ходом.

Что тут? Найти такие значениея, любой ход из которых приведет либо в выигрыш по первому ходу (**True в  функции PW1()**) **ИЛИ** выигрыш по второму ходу (**True в  функции PW2()**)


```python
def PP(s):
    return (PW1(s+1) or PW2(s+1)) and (PW1(s+2) or PW2(s+2)) and (PW1(s*2) or PW2(s*2)) 
```

Соберем всё в полных код

```python
# 1. Поменять условия выигрыша в PW1
# 2. Поменять операции исходя из ходов по задаче
# 3. Границы цикла
def PW1(s):
    return s+1>=38 or s+2>=38 or s*2>=38

def PP1(s):
    return PW1(s)!=True and (PW1(s+1) 
                             and PW1(s+2) and PW1(s*2))
    
def PW2(s):
    return PP1(s+1) or PP1(s+2) or PP1(s*2)

def PP(s):
    return (PW1(s+1) or PW2(s+1)) and (PW1(s+2) or PW2(s+2)) and (PW1(s*2) or PW2(s*2))  

for i in range(1,38):
    if PP(i): print(i) #Отсюда возьмем мин. знач
```
